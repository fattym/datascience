```python
import pandas
from sklearn.cluster import KMeans

wateTreatment = pandas.read_csv("https://archive.ics.uci.edu/ml/machine-learning-databases/water-treatment/water-treatment.data")
wateTreatment.columns=['id','Q-E','ZN-E','PH-E', 'DBO-E','DQO-E','SS-E', 'SSV-E', 'SED-E', 'COND-E', 'PH-P', 'DBO-P', 'SS-P' ,'SSV-P','SED-P',  'COND-P','PH-D','DBO-D','DQO-D', 'SS-D',  'SSV-D', 'SED-D', 'COND-D','PH-S','DBO-S','DQO-S','SS-S',  'SSV-S' ,'SED-S','COND-S','RD-DBO-P','RD-SS-P', 'RD-SED-P','RD-DBO-S','RD-DQO-S','RD-DBO-G', 'RD-DQO-G','RD-SS-G','RD-SED-G', ]         
wateTreatment.to_csv('data.csv',index=None,header=True)
wateTreatment
```

```python
wateTreatment_= pandas.read_csv('data.csv')
# wateTreatment_.isnull().sum()
# wateTreatment_.info()
# PH-E PH-P  PH-D COND-D  floats
wateTreatment_=wateTreatment_[['id','Q-E','ZN-E', 'DBO-E','DQO-E','SS-E', 'SSV-E', 'SED-E', 'COND-E', 'DBO-P', 'SS-P' ,'SSV-P','SED-P', 'COND-P','DBO-D','DQO-D', 'SS-D',  'SSV-D', 'SED-D', 'PH-S','DBO-S','DQO-S','SS-S',  'SSV-S' ,'SED-S','COND-S','RD-DBO-P','RD-SS-P', 'RD-SED-P','RD-DBO-S','RD-DQO-S','RD-DBO-G', 'RD-DQO-G','RD-SS-G','RD-SED-G','PH-E','PH-P', 'PH-D','COND-D',]]
 
wateTreatment_
```

```python
array=wateTreatment_.values
x=array[:,35:39]
from sklearn.cluster import KMeans
model=KMeans(n_clusters=5,random_state=42)# clasters into 5 people
model.fit(x)
```

```python
means= model.cluster_centers_
cluster = pandas.DataFrame(means,columns=['PH-E','PH-P', 'PH-D','COND-D',])
cluster
```

```python
wateTreatment_['label']=  model.labels_
wateTreatment_ =wateTreatment_[wateTreatment_['label']==3]
wateTreatment_
```
